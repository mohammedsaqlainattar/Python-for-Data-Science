## 3. Extract the rating as your target variable ‘y’ and all numerical parameters as your predictors ‘x’. Separate 25% of your data as test set.

```python
x = df.loc[:,'calories':'cups']
y = df['rating']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=10)
```

## 4. Fit a linear regression model and measure the mean squared error on test dataset.
### [ Hint: Explore linear models and metrics section of sklearn documentation]

```python
from sklearn import linear_model, metrics
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
print('Regression coefficients are:\n',model.coef_)
```

```python
model.intercept_
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
reg_df = pd.DataFrame({'Actual':y_test,'Predicted':y_pred,'Residuals':y_test-y_pred})
reg_df
```

## Mean squared error

```python
print('Mean Squared Error:',metrics.mean_squared_error(y_test,y_pred))
```

## Mean of the Residuals close to zero

```python
reg_df['Residuals'].mean()
```

## Correlation Matrix

```python
corrmat = df.corr()
corrmat
```

## Residuals vs Fitted values plot. The plot shows no patterns and residuals are evenly distributed. Homoscedasticity confirmed.

```python
plt.scatter(y_pred,reg_df['Residuals'])
plt.hlines(0,0,95,linestyles='--')
plt.ticklabel_format(style='plain',axis='y')
plt.xlabel('y_pred')
plt.ylabel('Residuals')
```

## Regression Score

```python
print('R^2 value:',metrics.r2_score(y_test,y_pred))
```

## Autocorrelation check (Durbin Watson test) - Value is 2.08 - No autocorrelation

```python
from statsmodels.stats.stattools import durbin_watson

durbin_watson(reg_df['Residuals'])
```
