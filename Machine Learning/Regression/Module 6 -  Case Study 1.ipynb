## 2.Data Manipulation:
## a.Create a new column -’total_benefitted’ that is a sum of inmates benefitted through all modes.[Hint: Use sum method with appropriate axis]
## b.Create a new row -“totals” that is the sum of all inmates benefitted through each mode across all states

### a.Create a new column -’total_benefitted’ that is a sum of inmates benefitted through all modes.[Hint: Use sum method with appropriate axis]

```python
df['total_benefitted'] = df.loc[:,'No. of Inmates benefitted by Elementary Education':'No. of Inmates benefitted by Computer Course'].sum(axis=1)
```

```python
df
```

### b.Create a new row -“totals” that is the sum of all inmates benefitted through each mode across all states

```python
df.append({'STATE/UT':'Totals',
           'YEAR':'',
           'No. of Inmates benefitted by Elementary Education':df['No. of Inmates benefitted by Elementary Education'].sum(),
           'No. of Inmates benefitted by Adult Education':df['No. of Inmates benefitted by Adult Education'].sum(),
           'No. of Inmates benefitted by Higher Education':df['No. of Inmates benefitted by Higher Education'].sum(),
           'No. of Inmates benefitted by Computer Course':df['No. of Inmates benefitted by Computer Course'].sum(),
           'total_benefitted':df['total_benefitted'].sum()}, ignore_index=True)
```

### a.Make a bar plot with each state name on the x -axis and their total benefitted inmates as their bar heights. Which state has the maximum number of beneficiaries?[Hint: Use bar method of pyplot]

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(12,8))

plt.bar(df['STATE/UT'],df['total_benefitted'],edgecolor='black',linewidth=1)
plt.title('Total no. of beneficiaries by State')
plt.ylabel('No. of beneficiaries')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

### Andhra Pradesh has the highest no. of beneficiaries
