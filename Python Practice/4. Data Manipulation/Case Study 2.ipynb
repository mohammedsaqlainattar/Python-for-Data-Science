## 1.Compute how much total salary cost has increased from year 2011 to 2014

```python
grouped_df_year = df.groupby(['Year']).agg({'TotalPay':'sum'})
```

```python
grouped_df_year
```

```python
salary_diff = grouped_df_year.loc[2014,'TotalPay'] - grouped_df_year.loc[2011,'TotalPay']
```

```python
print('The total salary cost increase from year 2011 to 2014 is:',round(salary_diff,2))
```

## 2.Which Job Title in Year 2014 has highest mean salary?

```python
filt = (df['Year'] == 2014)
df_salary_group = df.loc[filt].groupby('JobTitle').agg({'TotalPay':'mean'})
df_salary_group
```

```python
filt1 = (df_salary_group['TotalPay'] == df_salary_group['TotalPay'].max())
max_mean_salary = df_salary_group.loc[filt1]
max_mean_salary
```

```python
print('The Job title with the highest mean salary in the Year 2014 is:',max_mean_salary.index[0])
```

## 3.How much money could have been saved in Year 2014 by stopping OverTimePay?

```python
filt3 = (df['Year'] == 2014)
df.loc[filt3]
```

```python
grouped_df_year_savings = df.groupby(['Year']).agg({'OvertimePay':'sum'})
grouped_df_year_savings
```

```python
print('Total savings in Year 2014 by stopping OverTimePay is:',grouped_df_year_savings.loc[2014,'OvertimePay'])
```

## 4.Which are the top 5 common job in Year 2014 and how much do they cost SFO ?

```python
df_mostcommon_job = df.groupby(['JobTitle']).agg({'JobTitle':'count','TotalPay':'sum'})
```

```python
df_mostcommon_job.rename(columns={'JobTitle':'Count','TotalPay':'Cost to SFO'}, inplace=True)
```

```python
df_mostcommon_job
```

```python
df_mostcommon_job = df_mostcommon_job.sort_values(by='Count',ascending=False)
df_mostcommon_job
```
