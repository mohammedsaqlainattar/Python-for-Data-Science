## 1.Plot Total Sales Per Month for Year 2011.  How the total sales have increased over months in Year 2011. Which month has lowest Sales?

```python
filt = (df['Year'] == 2011)
data_2011 = df.loc[filt]
```

```python
data_2011
```

```python
monthly_group = data_2011.filter(items=['Month','Amount']).groupby(['Month'], as_index=False).sum()
monthly_group
```

```python
plt.style.use('seaborn')

plt.plot(monthly_group['Month'],monthly_group['Amount'], linewidth=2, marker='o', label='Sales')
plt.ticklabel_format(style='plain',axis='y')
plt.title('Total Sales vs Month (2011)')
plt.legend()
plt.xlabel('Month')
plt.ylabel('Amount')
plt.tight_layout()
plt.show()
```

### How the total sales haveincreased over months in Year 2011: Sales were fluctuating from 1st Month till 7th Month. Sales saw an upward trend from 8th Month till 11th Month.

### Which month has lowest Sales? : February

## 2.Plot Total Sales Per Month for Year 2011 as Bar Chart.  Is Bar Chart Better to visualize than Simple Plot?

```python
plt.style.use('ggplot')

plt.bar(monthly_group['Month'],monthly_group['Amount'], label='Sales')
plt.ticklabel_format(style='plain',axis='y')
plt.title('Total Sales vs Month (2011)')
plt.legend()
plt.xlabel('Month')
plt.ylabel('Amount')
plt.tight_layout()
plt.show()
```

### In my opinion line plots are better to visualize than bar chart in this case since then vartiation is not huge

# 3.Plot Pie Chart for Year 2011 Country Wise. Which Country contributes highest towards sales?

```python
country_group = data_2011.filter(items=['Country','Amount']).groupby('Country').sum()
country_group
```

```python
plt.style.use('seaborn')
fig, ax = plt.subplots()

ax.pie(country_group['Amount'], labels=country_group.index, wedgeprops = {'edgecolor':'black'});
ax.set_title('Pie Chart')
plt.tight_layout()
plt.show()
```

## The above Chart has too many items for a Pie Chart

## Top 5 Countries

```python
top_5 = country_group.nlargest(5,columns='Amount')
top_5
```

```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots()

ax.pie(top_5['Amount'], labels=top_5.index, wedgeprops = {'edgecolor':'black'},explode=[0.2,0,0,0,0], autopct='%1.2f%%', shadow=True)
ax.set_title('Pie Chart')
plt.tight_layout()
plt.show()
```

###  Which Country contributes highest towards sales? : United Kingdom

## 4.Plot Scatter Plot for the invoice amounts and see the concentration of amount.  In which range most of the invoice amounts are concentrated

```python
invoice_amount = data_2011.filter(items=['InvoiceDate','Amount']).groupby('InvoiceDate').sum()
invoice_amount
```

```python
plt.style.use('tableau-colorblind10')

plt.scatter(invoice_amount.index,invoice_amount['Amount'], s=100, edgecolor='black', linewidth=1, alpha=0.75)

plt.title('Scatter Plot')
plt.xlabel('InvoiceDate')
plt.ylabel('Amount')
plt.ylim((0,100000))
plt.tight_layout()

plt.show()
```

### In which range most of the invoice amounts are concentrated : Between 20k-40k
