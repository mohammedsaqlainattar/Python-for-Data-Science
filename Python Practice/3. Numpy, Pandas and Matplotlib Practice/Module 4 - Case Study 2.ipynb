## 1.Read the three csv files which contains the score of same students in term1 for each Subject

```python
import pandas as pd
```

```python
df_math = pd.read_csv(r"C:\Users\saqla\Desktop\Python Certification for Data Science\Modules\Module 4\Dataset\MathScoreTerm1.csv", index_col='ID')
df_ds = pd.read_csv(r"C:\Users\saqla\Desktop\Python Certification for Data Science\Modules\Module 4\Dataset\DSScoreTerm1.csv", index_col='ID')
df_physics = pd.read_csv(r"C:\Users\saqla\Desktop\Python Certification for Data Science\Modules\Module 4\Dataset\PhysicsScoreTerm1.csv", index_col='ID')
```

```python
df_ds
```

## 2.Remove the name and ethnicity column (to ensure confidentiality

```python
df_math.drop(columns=['Name','Ethinicity'], inplace=True)
df_ds.drop(columns=['Name','Ethinicity'], inplace=True)
df_physics.drop(columns=['Name','Ethinicity'], inplace=True)
```

```python
df_physics
```

## 3.Fill missing score data with zero

```python
df_math.fillna(0,inplace=True)
df_ds.fillna(0,inplace=True)
df_physics.fillna(0,inplace=True)
```

```python
df_physics.isna().sum()
```

## 4.Merge the three files

```python
merged_df = df_math.merge(df_ds, on='ID', suffixes=('_math','_ds')).merge(df_physics,on='ID', suffixes=('_ds','_physics'))
merged_df
```

```python
merged_df.drop(columns=['Age_math','Age_ds','Sex_math','Sex_ds'], inplace=True)
merged_df
```

```python
merged_df.rename(columns={'Score':'Score_physics','Subject':'Subject_physics'}, inplace=True)
merged_df
```

## 5.Change Sex(M/F) Columnto 1/2 for further analysis

```python
merged_df['Sex'].replace({'M':1,'F':2},inplace=True)
merged_df
```

## 6.Store the data in new file –ScoreFinal.csv

```python
merged_df.to_csv(r"C:\Users\saqla\Desktop\Python Certification for Data Science\Modules\Module 4\Dataset\ScoreFinal.csv")
```
